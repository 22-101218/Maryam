```python
df=pd.read_csv('Peer Pressure and Academic Performance (Responses) - Form Responses 1.csv')
```

```python
median_value = df['What is your CGPA?'].median()
condition = df['What is your CGPA?'] < 0
condition2 = df['What is your CGPA?'] == 0
df.loc[condition, 'What is your CGPA?'] = median_value
df.loc[condition2, 'What is your CGPA?'] = median_value
```

```python
plt.hist(df['On a scale of 1 to 5, how often do you feel pressured by your peers to perform well academically? '], bins=5, edgecolor='black')
plt.xlabel('Frequency of Feeling Pressured by Peers')
plt.ylabel('Frequency')
plt.title('Histogram of Feeling Pressured by Peers')
plt.show()
```

```python
plt.hist(df['On a scale of 1 to 5, how stressed do you feel about your academic performance due to peer pressure?'], bins=5, edgecolor='black')
plt.xlabel('Stress Levels due to Academic Performance')
plt.ylabel('Frequency')
plt.title('Histogram of Stress Levels due to Academic Performance')
plt.show()
```

```python
plt.boxplot(df['What is your CGPA?'])
plt.xlabel('CGPA')
plt.title('Box Plot of CGPA')
plt.show()
```

```python
plt.scatter(df['On a scale of 1 to 5, how often do you feel pressured by your peers to perform well academically? '], df['What is your CGPA?'], label='Peer Pressure Frequency', alpha=0.5)
plt.scatter(df['On a scale of 1 to 5, how stressed do you feel about your academic performance due to peer pressure?'], df['What is your CGPA?'], label='Stress Level', alpha=0.5)
plt.xlabel('Peer Pressure/Stress Level')
plt.ylabel('CGPA')
plt.title('Correlation between CGPA and Peer Pressure/Stress')
plt.legend()
plt.show()
```

```python
correlation = df['On a scale of 1 to 5, how often do you feel pressured by your peers to perform well academically? '].corr(df['What is your CGPA?'])
print("Correlation coefficient between Peer Pressure Frequency and CGPA:", correlation)
```

```python
sns.regplot(x='On a scale of 1 to 5, how often do you feel pressured by your peers to perform well academically? ', y='What is your CGPA?', data=df, line_kws={"color":"red"})
plt.xlabel('Peer Pressure Frequency')
plt.ylabel('CGPA')
plt.title('Simple Linear Regression: Peer Pressure Frequency vs. CGPA')
plt.show()
```
